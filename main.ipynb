## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
        ok_in_alphabet = []
        new_string = string.lower()
        alphabet = 'abcdefghijklmnopqrstuvwxyz'
        for abc in alphabet:
            if abc in new_string:
                ok_in_alphabet.append(True)
            else:
                ok_in_alphabet.append(False)
        if len(ok_in_alphabet) == len(alphabet):
            return True
        else:
            return False
        
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python
def sort_alpha(string):
    list_string = list(set(string))
    list_sort = list_string.sort(reverse=False)
    string_sort = ",".join(list_sort)
    
    return string_sort
    
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
def check_pass(string):
    ok_pass = 0
    not_pass = 0
    characters = ["#","@","!","$","%","&","( )","^","*","[ ]","{ }"]
    if len(string) >= 8:
        not_pass+=1
    else:
        not_pass+=0
    for s in string:
        if s == s.lower():
            ok_pass+=1
        else:
            not_pass+=0
    for s in string:
        if s == s.upper():
            ok_pass+=1
        else:
            not_pass+=0
    for s in string:
        if s in s.upper():
            ok_pass+=1
        else:
            not_pass+=0
    for s in string:
        if s in characters:
            ok_pass+=1
        else:
            not_pass+=0
            
    if ok_pass >= 4:
        return True
    else:
        return False      
```

```python
# This will test your function 
test_pass(check_pass)
```
